```python
import numpy as np
import random
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam
from sklearn.decomposition import PCA
```

```python
DATASET_PATH = "PlantVillage"

datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

train_data = datagen.flow_from_directory(
    DATASET_PATH,
    target_size=(128,128),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

val_data = datagen.flow_from_directory(
    DATASET_PATH,
    target_size=(128,128),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

NUM_CLASSES = train_data.num_classes
print("Total Classes:", NUM_CLASSES)
```

```python
def build_model(lr, filters, dropout):

    model = models.Sequential([
        layers.Conv2D(int(filters), (3,3), activation='relu', input_shape=(128,128,3)),
        layers.MaxPooling2D(2,2),

        layers.Conv2D(int(filters*2), (3,3), activation='relu'),
        layers.MaxPooling2D(2,2),

        layers.GlobalAveragePooling2D(),

        layers.Dense(128, activation='relu'),
        layers.Dropout(dropout),
        layers.Dense(NUM_CLASSES, activation='softmax')
    ])

    model.compile(
        optimizer=Adam(learning_rate=lr),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model
```

```python
class GWO:
    def __init__(self, n_wolves=3, max_iter=3):
        self.n_wolves = n_wolves
        self.max_iter = max_iter
        self.lb = np.array([0.0001, 16, 0.2])
        self.ub = np.array([0.01, 64, 0.5])

    def fitness(self, position):
        model = build_model(position[0], int(position[1]), position[2])

        history = model.fit(
            train_data,
            epochs=2,
            validation_data=val_data,
            verbose=0
        )

        return history.history['val_accuracy'][-1]

    def optimize(self):

        wolves = np.random.uniform(self.lb, self.ub, (self.n_wolves, 3))

        for iteration in range(self.max_iter):

            print(f"\nIteration {iteration+1}")

            fitness_scores = []

            for i in range(self.n_wolves):
                fitness = self.fitness(wolves[i])
                fitness_scores.append(fitness)

            sorted_indices = np.argsort(fitness_scores)[::-1]

            alpha = wolves[sorted_indices[0]].copy()
            beta  = wolves[sorted_indices[1]].copy()
            delta = wolves[sorted_indices[2]].copy()

            a = 2 - iteration * (2 / self.max_iter)

            for i in range(self.n_wolves):

                for j in range(3):

                    r1, r2 = np.random.rand(), np.random.rand()
                    A1 = 2 * a * r1 - a
                    C1 = 2 * r2
                    D_alpha = abs(C1 * alpha[j] - wolves[i][j])
                    X1 = alpha[j] - A1 * D_alpha

                    r1, r2 = np.random.rand(), np.random.rand()
                    A2 = 2 * a * r1 - a
                    C2 = 2 * r2
                    D_beta = abs(C2 * beta[j] - wolves[i][j])
                    X2 = beta[j] - A2 * D_beta

                    r1, r2 = np.random.rand(), np.random.rand()
                    A3 = 2 * a * r1 - a
                    C3 = 2 * r2
                    D_delta = abs(C3 * delta[j] - wolves[i][j])
                    X3 = delta[j] - A3 * D_delta

                    wolves[i][j] = (X1 + X2 + X3) / 3

                wolves[i] = np.clip(wolves[i], self.lb, self.ub)

        return alpha
```

```python
gwo = GWO(n_wolves=3, max_iter=3)
best_params = gwo.optimize()

print("\nBest Parameters Found:")
print("Learning Rate:", best_params[0])
print("Filters:", int(best_params[1]))
print("Dropout:", best_params[2])
```

```python
final_model = build_model(
    best_params[0],
    int(best_params[1]),
    best_params[2]
)

history = final_model.fit(
    train_data,
    epochs=10,
    validation_data=val_data
)
```

```python
feature_model = tf.keras.Model(
    inputs=final_model.inputs,
    outputs=final_model.layers[-3].output
)

features = feature_model.predict(train_data)
```

```python
pca = PCA(n_components=50)

features_pca = pca.fit_transform(features)

print("Original Features:", features.shape)
print("After PCA:", features_pca.shape)
```
